```python
import pandas as pd
data = pd.read_csv("used_cars_data.csv")
data[data.duplicated(keep=False)]
print(data)
```

```python
data=data.drop(['S.No.'],axis=1)
data.info()
```

```python
# Extract numeric values from 'Mileage' and 'Engine' columns
data['Mileage_Clean'] = data['Mileage'].str.extract(r'([\d.]+)').astype(float)
data['Engine_Clean'] = data['Engine'].str.extract(r'(\d+)').astype(float)

# Create new feature: Mileage per CC
data['Mileage_per_CC'] = data['Mileage_Clean'] / data['Engine_Clean']
data.info()

```

```python
# Add the feature
data['Is_First_Owner'] = data['Owner_Type'].apply(lambda x: 1 if str(x).strip().lower() == 'first' else 0)

# View all columns in the first few rows
print(data.head())
```

```python
data.drop(columns=['Mileage_Clean', 'Engine_Clean', 'Mileage_per_CC'], inplace=True, errors='ignore')
data.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("used_cars_data.csv")

# Encode categorical variables to numeric
df['Fuel_Code'] = pd.factorize(df['Fuel_Type'])[0]
df['Name_Code'] = pd.factorize(df['Name'])[0]

# Plot scatter
plt.figure(figsize=(10, 6))
plt.scatter(df['Fuel_Code'], df['Name_Code'], alpha=0.5)
plt.xlabel('Fuel Type (Encoded)')
plt.ylabel('Car Model (Encoded)')
plt.title('Scatter Plot: Fuel Type vs Car Model')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("used_cars_data.csv")

df['Mileage'] = df['Mileage'].str.extract(r'(\d+\.\d+|\d+)').astype(float)


df = df[['Price', 'Mileage']].dropna()

plt.figure(figsize=(8, 5))
plt.scatter(df['Mileage'], df['Price'], alpha=0.5, color='orange')
plt.xlabel('Mileage (kmpl or km/kg)')
plt.ylabel('Price (in lakhs)')
plt.title('Scatter Plot: Price vs Mileage')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("used_cars_data.csv")


df['Brand'] = df['Name'].str.split().str[0]


brand_location_counts = df.groupby(['Brand', 'Location']).size().unstack(fill_value=0)

plt.figure(figsize=(12, 6))
brand_location_counts.plot(kind='bar', stacked=False, figsize=(12, 6))
plt.title('Car Brand vs Location')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("used_cars_data.csv")

plt.figure(figsize=(8, 5))
sns.countplot(x='Fuel_Type', data=data, color='grey', order=data['Fuel_Type'].value_counts().index)
plt.title('Count of Cars by Fuel Type')
plt.xlabel('Fuel Type')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
